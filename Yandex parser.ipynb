```python
import time
```

```python
new_global_list = []

for link in mylist:
    
    try:
        driver.get(link)
    except:
        print('Doesn`t work', link)
        continue
        
    html = BeautifulSoup(driver.page_source)
    
    print('Link' + ' ' + 'accepted')
        
    buffer= []
    
    #Название объявления
    
    dirty = html.find_all('h1',attrs={'class':'OfferHeader__title'})
    for i in dirty:
        buffer.append(i.text)
    
    #времядо метро
    dirty1 = html.find_all('span',attrs={'class':'MetroWithTime__distance OfferHeaderLocation__station-distance'})
    if len(dirty1) > 0:
        buffer.append(dirty1[0].text)
    else:
        buffer.append('NA')
        
    #станция метро
    dirty2 = html.find_all('span',attrs={'class':'MetroStation__title'})
    if len(dirty2) > 0:
        buffer.append(dirty2[0].text)
    else:
        buffer.append('NA')
            
         
    #Этаж, адрес, цена ₽ за м²
    
    dirty3 = html.find_all('div',attrs={'class':'OfferCard__text'})
    
    if len(dirty3) > 0:
        buffer.append(dirty3[0].text)
        buffer.append(dirty3[1].text)
        buffer.append(dirty3[2].text)
    else:
        buffer.append('NA')
    
    #Удобства
    dirty4 = html.find_all('span',attrs={'class':'OfferDetailsDescription__feature'})
    buffer2 = []
    for i in dirty4:
        if len(i) > 0:
            buffer2.append(i.text)
        else:
            buffer2.append('NA')
            
    buffer.append(buffer2)
   
    
    #Price
    
    dirty5 = html.find_all('span',attrs={'class':'price'})
    if len(dirty5) > 0:
        buffer.append(dirty5[0].text)
    else:
        buffer.append('NA')
    
    
    new_global_list.append(buffer)
    time.sleep(3)ОШШШШШШШШЧЧЛМИ
```
